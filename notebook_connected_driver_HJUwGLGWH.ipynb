```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pandas as pd

from bokeh.plotting import figure, output_notebook, show, output_file
from bokeh.models.widgets import Panel, Tabs,Div, DateSlider
from bokeh.models import ColumnDataSource, HoverTool, BoxAnnotation, Span
from bokeh.layouts import gridplot, column, row, layout
from bokeh.palettes import Spectral6
from bokeh.transform import factor_cmap
from bokeh.core.properties import value

from datetime import datetime as dt
import time

import numpy as np
output_notebook()
```

```python
df_data_1 = pd.read_csv("https://raw.githubusercontent.com/jrartd/jorgerobalino.co/master/enero-febrero.csv", sep=";")
df_data_1.head()

```

```python
df_data_2 = pd.read_csv("https://raw.githubusercontent.com/jrartd/jorgerobalino.co/master/mayo-junio.csv", sep=";", encoding = "ISO-8859-1")
df_data_2.head()
```

```python
data_raw = df_data_2
data_rawdos = df_data_1
```

# Mayo - Junio  

```python
#conversiones data_raw
data_raw["driving_score"] = pd.to_numeric(data_raw["driving_score"],errors='coerce' )
data_raw["max_speed"] = pd.to_numeric(data_raw["max_speed"],errors='coerce' )
data_raw["harsh_acceleration"] = pd.to_numeric(data_raw["harsh_acceleration"],errors='coerce' )
data_raw["harsh_braking"] = pd.to_numeric(data_raw["harsh_braking"],errors='coerce' )
data_raw["over_speed"] = pd.to_numeric(data_raw["harsh_braking"],errors='coerce' )
data_raw["harsh_braking_turn"] = pd.to_numeric(data_raw["harsh_braking"],errors='coerce' )
data_raw['entry_time']= pd.to_datetime(data_raw['entry_time'], errors='coerce') 

#nuevo dataframe
datadriver = data_raw[["entry_time","imei","driving_score","over_speed","max_speed","harsh_acceleration","harsh_braking","harsh_braking_turn"]]
datadriver["Estado"] = "despues"
datadriver = datadriver.drop([0],axis=0)
datadriver.head()
```

# Enero - Febrero

```python
data_rawdos["driving_score"] = pd.to_numeric(data_rawdos["driving_score"],errors='coerce' )
data_rawdos["max_speed"] = pd.to_numeric(data_rawdos["max_speed"],errors='coerce' )
data_rawdos["harsh_braking_turn"] = pd.to_numeric(data_rawdos["harsh_acceleration"],errors='coerce' )
data_rawdos["harsh_acceleration"] = pd.to_numeric(data_rawdos["harsh_acceleration"],errors='coerce' )
data_rawdos["over_speed"] = pd.to_numeric(data_rawdos["harsh_acceleration"],errors='coerce' )
data_rawdos["harsh_braking"] = pd.to_numeric(data_rawdos["harsh_braking"],errors='coerce' )


#cadena
#data_raw["imei"] = data_raw["imei"].to_string()

#fecha
data_rawdos['entry_time']= pd.to_datetime(data_rawdos['entry_time'], errors='coerce') 

datadriverdos  = data_rawdos[["entry_time","imei","driving_score","over_speed","max_speed","harsh_acceleration","harsh_braking","harsh_braking_turn"]]
datadriverdos["Estado"] = "antes"

datadriverdos.head()
```

```python
df_all_rows = pd.concat([datadriver, datadriverdos])
df_all_rows.head()
```

# Global

```python
# Stacked

fruits = ["driving_score","over_speed","max_speed","harsh_acceleration","harsh_braking","harsh_braking_turn"]
years = ["Antes", "Despues"]
colors = ["#ffb300", "#ff8400"]

data = {'fruits' : fruits,
        'Antes'   : [datadriverdos["driving_score"].mean(),datadriverdos["over_speed"].mean(),datadriverdos["max_speed"].mean(),datadriverdos["harsh_acceleration"].mean(),datadriverdos["harsh_braking"].mean(),datadriverdos["harsh_braking_turn"].mean()],
        'Despues'   : [datadriver["driving_score"].mean(),datadriverdos["over_speed"].mean(),datadriver["max_speed"].mean(),datadriver["harsh_acceleration"].mean(),datadriver["harsh_braking"].mean(),datadriver["harsh_braking_turn"].mean()]
       }

p1 = figure(x_range=fruits,plot_width=1000, plot_height=400, title="Comparativa entre los promedios Enero - Febrero y Mayo-Junio", tools="")

p1.vbar_stack(years, x='fruits', width=0.9, color=colors, source=data,
             legend=[value(x) for x in years])

p1.add_tools(HoverTool(
        tooltips=[
            ( 'Antes',  '@{Despues}' ), # use @{ } for field names with spaces
            ( 'Despues',  '@{Antes}' ), # use @{ } for field names with spaces
            

        ],

        formatters={
            'date'      : 'datetime', # use 'datetime' formatter for 'date' field
            'adj close' : 'printf',   # use 'printf' formatter for 'adj close' field
                                      # use default 'numeral' formatter for other fields
        },

        
))


p1.y_range.start = 0
p1.x_range.range_padding = 0.1
p1.xgrid.grid_line_color = None
p1.axis.minor_tick_line_color = None
p1.outline_line_color = None
p1.legend.location = "top_left"
p1.legend.orientation = "horizontal"
p1.legend.click_policy="hide"

show(p1)
```
